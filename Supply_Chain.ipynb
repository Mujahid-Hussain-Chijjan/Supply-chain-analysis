# ANALYSIS

## 1. Late delivery by Order Region, Product Category and Customer segment and Late delivery Risk Analysis

### 1.1 Late delivery count by Order Region

#### Decriptive Analysis for Late delivery by Order Region

Calculating Late delivery percentage.

```python
late_deliveries = dfsupply_chain[dfsupply_chain['Delivery Status'] == 'Late delivery']
late_delivery_rate = len(late_deliveries) / len(dfsupply_chain) * 100
print('Late delivery rate:', late_delivery_rate)
```

1st code shows the count of Late deliveries by Order Region.
2nd code show percentage of Late deliveries by Order Region.

```python
late_deliveries_by_region = dfsupply_chain[dfsupply_chain['Delivery Status'] == 'Late delivery']['Order Region'].value_counts()
print("Late Deliveries by Order Region:")
print(late_deliveries_by_region)
```

```python
total_late_deliveries = dfsupply_chain['Delivery Status'].value_counts()['Late delivery']
late_deliveries_by_region = dfsupply_chain[dfsupply_chain['Delivery Status'] == 'Late delivery']['Order Region'].value_counts()
late_deliveries_by_region_pct = late_deliveries_by_region / total_late_deliveries * 100
print("Late Deliveries by Order Region:")
print(late_deliveries_by_region_pct)
```

#### Visuals of Late delivery by Order Region.

```python
plt.figure(figsize=(12, 6))
bars = late_deliveries_by_region.plot(kind='bar', color='blue')
plt.title('Count of Late Deliveries by Order Region')
plt.xlabel('Order Region')
plt.ylabel('Count of Late Deliveries')
for bar in bars.patches:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), round(bar.get_height(), 2), ha='center', va='bottom')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
bars = late_deliveries_by_region_pct.plot(kind='bar', color='Skyblue')
plt.title('Percentage of Late Deliveries by Order Region')
plt.xlabel('Order Region')
plt.ylabel('Percentage of Late Deliveries')
for bar in bars.patches:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), round(bar.get_height(), 2), ha='center', va='bottom')
plt.show()
```

### 3.2 Product Analysis by Region

#### Descriptive Analysis

```python
product_sales_by_region = dfsupply_chain.groupby(['Order Region', 'Product Name'])['Sales'].sum().reset_index()
highest_sales_by_region = product_sales_by_region.groupby('Order Region').apply(lambda x: x.loc[x['Sales'].idxmax()])
for region in dfsupply_chain['Order Region'].unique():
    print(f"\nRegion: {region}")
    highest_sales_product = highest_sales_by_region.loc[region, 'Product Name']
    highest_sales_value = highest_sales_by_region.loc[region, 'Sales']
    print(f"Highest sales product: {highest_sales_product} with total sales of {highest_sales_value}")
```

```python
product_sales_by_region = dfsupply_chain.groupby(['Order Region', 'Product Name'])['Sales'].sum().reset_index()
least_sales_by_region = product_sales_by_region.groupby('Order Region').apply(lambda x: x.loc[x['Sales'].idxmin()])
for region in dfsupply_chain['Order Region'].unique():
    print(f"\nRegion: {region}")
    least_sales_product = least_sales_by_region.loc[region, 'Product Name']
    least_sales_value = least_sales_by_region.loc[region, 'Sales']
    print(f"Least sales product: {least_sales_product} with total sales of {least_sales_value}")
```

#### Visuals

```python
colormap = plt.cm.tab20b(np.linspace(0, 1, len(highest_sales_by_region)))
plt.figure(figsize=(25, 19))
bars = plt.bar(highest_sales_by_region.index, highest_sales_by_region['Sales'], alpha=0.7)
for bar, sale, product in zip(bars, highest_sales_by_region['Sales'], highest_sales_by_region['Product Name']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{round(sale, 2)}', ha='center', va='bottom')
plt.title('Product with Highest Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.show()
```
